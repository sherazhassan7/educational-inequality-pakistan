```python
import pandas as pd

sr = pd.read_csv('../Data:/school_resources.csv')

# Filter to Punjab government schools
punjab_govt = sr[(sr['province'] == 'Punjab') & (sr['school_type'] == 'Government')]
print(f"Punjab government schools: {len(punjab_govt)}")

# National Punjab textbook rate (paper claims govt national is 17.9%)
national_rate = punjab_govt['class2_has_textbooks'].mean() * 100
print(f"\nPunjab govt textbook rate (province-wide): {national_rate:.1f}%")
print(f"(Compare to overall national govt rate of 17.9% reported in paper)")

# Per-district breakdown
by_district = punjab_govt.groupby('district')['class2_has_textbooks'].mean().sort_values()
print(f"\nNumber of Punjab districts: {len(by_district)}")
print(f"\nDistrict-level distribution:")
print(f"  Min:    {by_district.min()*100:.1f}%")
print(f"  25th %: {by_district.quantile(0.25)*100:.1f}%")
print(f"  Median: {by_district.median()*100:.1f}%")
print(f"  75th %: {by_district.quantile(0.75)*100:.1f}%")
print(f"  Max:    {by_district.max()*100:.1f}%")

print(f"\nAll Punjab districts (sorted):")
for d, v in by_district.items():
    print(f"  {d}: {v*100:.1f}%")
```

```python
import pandas as pd
sr = pd.read_csv('../Data:/school_resources.csv')

punjab_govt = sr[(sr['province'] == 'Punjab') & (sr['school_type'] == 'Government')]
all_govt    = sr[sr['school_type'] == 'Government']

print("Metric                          | Punjab Govt | National Govt")
print("-" * 65)

# Textbooks (already verified)
print(f"Class 2 textbooks present       | {punjab_govt['class2_has_textbooks'].mean()*100:>9.1f}% | {all_govt['class2_has_textbooks'].mean()*100:>9.1f}%")

# Pupil-teacher ratio
print(f"Median pupil-teacher ratio      | {punjab_govt['pupil_teacher_ratio'].median():>10.1f} | {all_govt['pupil_teacher_ratio'].median():>10.1f}")

# Teacher attendance
print(f"Teacher attendance rate         | {punjab_govt['teacher_attendance_rate'].mean():>10.1f}% | {all_govt['teacher_attendance_rate'].mean():>10.1f}%")

# Student attendance  
print(f"Student attendance rate         | {punjab_govt['attendance_rate'].mean():>10.1f}% | {all_govt['attendance_rate'].mean():>10.1f}%")

# Blackboard
print(f"Class 2 blackboard present      | {punjab_govt['class2_has_blackboard'].mean()*100:>9.1f}% | {all_govt['class2_has_blackboard'].mean()*100:>9.1f}%")

# Supplementary material
print(f"Class 2 supplementary material  | {punjab_govt['class2_has_supplementary_material'].mean()*100:>9.1f}% | {all_govt['class2_has_supplementary_material'].mean()*100:>9.1f}%")
```
